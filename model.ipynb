```python
import numpy as np
import pandas as pd
import glob 
from skimage import io
import matplotlib.pyplot as plt 
from sklearn.preprocessing import Normalizer
from sklearn.decomposition import PCA
import pickle as pk
from PIL import Image
import cv2
```

```python
def model(img,imgs):
    img = img.convert('L')
    img_1= imgs[0].convert('L')
    img_2= imgs[1].convert('L')
    img_3 =imgs[2].convert('L')
    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    img=np.array(img)
    img_1=np.array(imgs[0])
    img_2=np.array(imgs[1])
    img_3=np.array(imgs[2])
    face= face_cascade.detectMultiScale(img, 1.3, 1)[0]
    face1 = face_cascade.detectMultiScale(img_1, 1.3, 1)[0]
    face2 = face_cascade.detectMultiScale(img_2, 1.3, 1)[0]
    face3 = face_cascade.detectMultiScale(img_3, 1.3, 1)[0]
    (x,y,w,h)=face
    img = img[y:y + h, x:x + w]
    cv2.imwrite('detcted.jpg', img)
    cv2.imshow('img', img)
    cv2.waitKey()
    (x,y,w,h)=face1
    img_1 = img_1[y:y + h, x:x + w]
    cv2.imwrite('detcted.jpg', img_1)
    cv2.imshow('img', img_1)
    cv2.waitKey()
    (x,y,w,h)=face2
    img_2= img_2[y:y + h, x:x + w]
    cv2.imwrite('detcted.jpg', img_2)
    cv2.imshow('img', img_2)
    cv2.waitKey()
    (x,y,w,h)=face3
    img_3= img_3[y:y + h, x:x + w]
    cv2.imwrite('detcted.jpg', img_3)
    cv2.imshow('img', img_3)
    cv2.waitKey()
    newsize=(243, 320)
    img= np.resize(img,(243,320))
    img_1=np.resize(img_1,(243,320))
    img_2=np.resize(img_2,(243,320))
    img_3=np.resize(img_3,(243,320))
    reshaped_img=img.reshape(1,img.shape[0]*img.shape[1])
    reshaped_img1=img_1.reshape(1,img_1.shape[0]*img_1.shape[1])
    reshaped_img2=img_2.reshape(1,img_2.shape[0]*img_2.shape[1])
    reshaped_img3=img_3.reshape(1,img_3.shape[0]*img_3.shape[1])
    reshaped_img1 = pd.DataFrame(reshaped_img1)
    reshaped_img2 = pd.DataFrame(reshaped_img2)
    reshaped_img3 = pd.DataFrame(reshaped_img3)
    reshaped_img = pd.DataFrame(reshaped_img)
    scalar = Normalizer()
    imgs_scaled1 = pd.DataFrame(scalar.fit_transform(reshaped_img1), columns=reshaped_img1.columns)
    imgs_scaled2 = pd.DataFrame(scalar.fit_transform(reshaped_img2), columns=reshaped_img2.columns)
    imgs_scaled3 = pd.DataFrame(scalar.fit_transform(reshaped_img3), columns=reshaped_img3.columns)
    img_scaled= pd.DataFrame(scalar.fit_transform(reshaped_img), columns=reshaped_img.columns)
    pca_reload = pk.load(open("pca1.pkl",'rb'))
    img_scaled= pca_reload.transform(img_scaled)
    imgs_scaled1= pca_reload.transform(imgs_scaled1)
    imgs_scaled2= pca_reload.transform(imgs_scaled2)
    imgs_scaled3= pca_reload.transform(imgs_scaled3)
    distance1=(np.sqrt(np.sum((img_scaled-imgs_scaled1)**2)))
    distance2=(np.sqrt(np.sum((img_scaled-imgs_scaled2)**2)))
    distance3=(np.sqrt(np.sum((img_scaled- imgs_scaled3)**2)))
    near=min(distance1,distance2,distance3)
    y_pred=int(near<0.04)
    return y_pred
```
